```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  
        
        
# Driver code     

def get_the_co(my_string):
    string_to_list=str(my_string)
    list_index=string_to_list.index("(")
    batas=len(string_to_list)
    list_cop=[]
    total_for_plot=[]
    for i in range(list_index,batas):
        list_cop.append(string_to_list[i])
    total_for_plot.append(listToString(list_cop))
    return total_for_plot

data_list=list(data['appears on'])
[item for item in data_list if item.startswith('(C')]
        
    
```
